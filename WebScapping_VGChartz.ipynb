```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def extraer_ventas_vgchartz_raw(total_paginas=50):
    datos_juegos = []
    print(f"Extracción de {total_paginas} páginas en VGChartz...")

    for pagina in range(1, total_paginas + 1):
        url = f"https://www.vgchartz.com/games/games.php?page={pagina}&results=200&order=Sales&dir=DESC"
        # Hacernos pasar por un navegador para evitar bloqueos
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
        respuesta = requests.get(url, headers=headers)

        soup = BeautifulSoup(respuesta.text, 'html.parser')

        for fila in soup.find_all('tr'):
            columnas = fila.find_all('td')

            # Verificar que sea una fila de tabla completa
            if len(columnas) >= 10:
                nombre = columnas[2].text.strip()
                ventas_texto = columnas[8].text.strip()

                # Solo tomamos la fila si la columna de ventas termina en m
                if ventas_texto.endswith('m'):
                    try:
                        # Convertimos el texto "5.20" a un número real
                        ventas_num = float(ventas_texto.replace('m', ''))

                        # Guardamos solo el nombre y las ventas
                        datos_juegos.append({
                            'nombre': nombre,
                            'ventas_globales_millones': ventas_num
                        })
                    except ValueError:
                        pass

        print(f"Página {pagina}/{total_paginas}.")
        time.sleep(1) # Pausa para evitar saturar el servidor

    return pd.DataFrame(datos_juegos)


df_ventas_raw = extraer_ventas_vgchartz_raw(total_paginas=50)


print(f"Total de registros limpios obtenidos: {len(df_ventas_raw)}")

# Guardar
df_ventas_raw.to_csv("dataset_vgchartz_ventas.csv", index=False, encoding='utf-8')
```
